```python
def add_column_prefix (df, column, prefix):
    return df[column].apply(lambda x: prefix + str(x))

df['Destination'] = add_column_prefix(df,'Destination', 'Dest_' )
df['Source'] = add_column_prefix(df,'Source', 'Sour_' )
```

```python
def onehot_encoder(df, columns):
    for column in columns:
        dummies = pd.get_dummies(df[column])
        df = pd.concat([df, dummies], axis = 1)
        df.drop(column, axis = 1 , inplace = True)
        
    return df
```

```python
cat_feat = ['Airline','Source','Destination',]

df = onehot_encoder(df,cat_feat )

```

```python
# Removing Outliers with Z-score

from scipy.stats import zscore

z_score = zscore(df[['Price', 'Duration_hours']])
abs_z_score = np.abs(z_score)

filtering_entry = (abs_z_score < 3).all(axis=1)

df= df[filtering_entry]
```

```python
# Data Scaling Using RobustScaler
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()

x = pd.DataFrame(scaler.fit_transform(x), columns = x.columns)
```

```python
x
```

```python
# Importing Linear Regression model for getting best random state 
from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
maxr2 = 0
maxRS=0

for i in range(0,500):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25, random_state=i)
    LR = LinearRegression()
    LR.fit(x_train,y_train)
    predf=LR.predict(x_test)
    r2_sc = r2_score(y_test, predf)
    
    if r2_sc>maxr2:
        maxr2=r2_sc
        macRS = i
print('Best R2score=',maxr2, "Random state=",macRS) 
```

```python
# Spitting data into training and testing dataset
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25, random_state=358)
```

```python
y_train
```

```python
# Going for Linear Regression model
LR = LinearRegression()

LR.fit(x_train,y_train)
predf=LR.predict(x_test)
r2_sc = r2_score(y_test, predf)
r2_sc
```

```python
# Checking with KNN model
from sklearn.neighbors import KNeighborsRegressor
KN = KNeighborsRegressor()
KN.fit(x_train,y_train)

predf=KN.predict(x_test)
r2_sc = r2_score(y_test, predf)
r2_sc
```

```python
# Checking for DecissionTreeRegressor
from sklearn.tree import DecisionTreeRegressor

DT =  DecisionTreeRegressor()

DT.fit(x_train,y_train)

predf=DT.predict(x_test)
r2_sc = r2_score(y_test, predf)
r2_sc
```

```python
# Checking for AdaBoostRegressor
from sklearn.ensemble import AdaBoostRegressor
AD =  AdaBoostRegressor()
AD.fit(x_train,y_train)

predf=AD.predict(x_test)
r2_sc = r2_score(y_test, predf)
r2_sc
```

```python
# Checking for RandomForest
from sklearn.ensemble import RandomForestRegressor
RF = RandomForestRegressor()

RF.fit(x_train,y_train)

predf=RF.predict(x_test)
r2_sc = r2_score(y_test, predf)
r2_sc
```

```python
import xgboost as xgb
xgb= xgb.XGBRegressor()
xgb.fit(x_train,y_train)
predf=xgb.predict(x_test)
r2_sc = r2_score(y_test, predf)
r2_sc
```

```python
from sklearn.model_selection import cross_val_score

scr = cross_val_score(LR, x,y, cv=5)
print("Cross validation score for Linear Regression is", scr.mean())

scr = cross_val_score(KN, x,y, cv=5)
print("Cross validation score for KNN is", scr.mean())


scr = cross_val_score(DT, x,y, cv=5)
print("Cross validation score for DecisionTreeRegressor is", scr.mean())

scr = cross_val_score(RF, x,y, cv=5)
print("Cross validation score for RANDOM FOREST is", scr.mean())

scr = cross_val_score(AD, x,y, cv=5)
print("Cross validation score for ADa boost is", scr.mean())

scr = cross_val_score(xgb, x,y, cv=5)
print("Cross validation score for XG Bootst Regressor is", scr.mean())

```
